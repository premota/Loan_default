# import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
data = pd.read_csv("Clean_data.csv")
```

```python
data.head(3)
```

# Explore numerical features

```python
numerical_df = data.select_dtypes(exclude = "object")
```

```python
numerical_df.head(2)
```

**Plot boxplot**

```python
def box_plot(df):
    num_features = len(df.columns)
    num_rows = (num_features +1)//2

    fig, axes = plt.subplots(ncols=2, nrows=num_rows, figsize = (10,15))
    axes = axes.flatten()

    for index, feature in enumerate(df.columns):
        ax = axes[index]
        sns.boxplot(df[feature], ax=ax, orient = 'h')
        plt.title(f"{feature}")

    plt.suptitle("Box plot of numerical features")
    plt.tight_layout()
    plt.show()
```

```python
box_plot(numerical_df)
```

**Plot correlation**

```python
correlation = numerical_df.corr()

mask = np.triu(np.ones_like(numerical_df.corr(), dtype=bool))
plt.figure(figsize = (20,6))
sns.heatmap(correlation, cmap = "viridis",mask = mask)
plt.xticks(rotation=45)
plt.show();
```
